```python
import requests
response = requests.get("https://gorest.co.in/public/v2/users")
response
```

```python
 data = response.json()
data
```

```python
data_rows = []
map_status_to_value = {"active":True,"inactive":False } #is_active

map_gender = {"male":1 ,"female": 0} #is_male
for single_dictionary in data:
    user_id = single_dictionary["id"]
    user_name = single_dictionary["name"]    
    user_email = single_dictionary["email"]    
    gender = map_gender[single_dictionary["gender"]]    
    status = map_status_to_value[single_dictionary["status"]]  
    data_rows.append([user_id, user_name,user_email,gender ,status])
data_rows[0]
```

```python
import csv
with open("shehab.csv", mode="w",encoding="utf-8",newline="") as f:
    writer = csv.writer(f)
    header = ["user_id", "user_name" , "user_email","is_male" ,"is_active" ]
    writer.writerow(header)
    writer.writerows(data_rows)
    
import pandas as pd

df = pd.read_csv("shehab.csv")
df
```

```python
import requests
response = requests.get("https://gorest.co.in/public/v2/posts")
data = response.json()
data_rows = []
for signla_dictionary in data:
    my_id = signla_dictionary["id"]
    user_id = signla_dictionary["user_id"]
    title = signla_dictionary["title"]
    body = signla_dictionary["body"]
    data_rows.append([my_id,user_id ,title ,body])
data_rows    
    
import csv
with open("my_work.csv", mode="w" ,encoding="utf_8", newline="") as f:
    writer = csv.writer(f)
    header = ["my_id","user_id","title","body"]
    writer.writerow(header)
    writer.writerows(data_rows)
import pandas as pd
df = pd.read_csv("my_work.csv")
df
```

```python
all_data = []
page = 0

while True:
    html = requests.get(f"https://wuzzuf.net/search/jobs/?a=navbl&q=data%20science%20jobs&start={page}") 
    soup = BeautifulSoup(html.text, "html.parser")
        
    all_div = soup.find_all("div", {"class":"css-pkv5jc"})
    if len(all_div) == 0:
        print("finished")
        break
    for div in all_div:
        title = div.find("a", {"class":"css-o171kl"}).text.strip()
        link = "https://wuzzuf.net" + all_div[0].find("a", {"class":"css-o171kl"}).attrs["href"].strip()
        location = div.find("span", {"class":"css-5wys0k"}).text.strip()
        skills  = div.find("div", {"class":"css-y4udm8"}).text.strip()        
        company  = div.find("a", {"class":"css-17s97q8"}).text[:-2].strip()
        all_data.append([title ,link ,location, skills,company])
    print(f"scraping page {page}")
    page += 1
```

```python
import csv
with open("jopes.csv", mode='w', encoding='utf_8', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['title','link','location' ,'skills' ,'company'])
    writer.writerows(all_data)
    
import pandas as pd
df = pd.read_csv("jopes.csv")
df
```
